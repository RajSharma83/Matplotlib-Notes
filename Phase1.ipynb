line chart -> plot()

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [18, 23, 22, 24]

plt.plot(x, y)        # Create line plot
plt.xlabel("X-axis")  # Label for X-axis
plt.ylabel("Y-axis")  # Label for Y-axis
plt.title("Line Plot")# Title of the plot
plt.show()            # Display the plot
```

```python
# Graph for Bakery Sales
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
sales = [150, 100, 160, 200, 250, 300, 120]
plt.plot(days, sales)
plt.xlabel("Days of the Week")
plt.ylabel("Number of Baked Goods Sold")
plt.title("Bakery Sales Over a Week")
plt.show()
```

```python
# plt.plot(x, y color='color_name' linestyle='line_Style'  marker='marker symbol' linewidth=value label='label name')
# plt.xlabel("X-axis")  # Label for X-axis
# plt.ylabel("Y-axis")  # Label for Y-axis

month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.plot(month, budget, color='blue', linestyle='--', marker='o' ,linewidth=2, label= 'Monthly Budget')
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.title("Monthly Budget Over 4 Months")
plt.show()
```

Legend() # Labels for different plots in a figure.
           Positions: 'upper right', 'lower left'.

```python
month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.plot(month, budget, color='blue', linestyle='--', marker='o' ,linewidth=2, label= 'Monthly Budget')
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.title("Monthly Budget Over 4 Months")
plt.legend(loc='upper left')  # at the upper left corner
# plt.legend(loc='lower right') # at the lower right corner
plt.show()
```

Grid() # adds grid lines to a plot to make it easier to read.

```python
month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.plot(month, budget, color='blue', linestyle='--', marker='o' ,linewidth=2, label= 'Monthly Budget')
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.title("Monthly Budget Over 4 Months")
plt.legend(loc='upper left')  # at the upper left corner
plt.grid(color='gray', linestyle=':', linewidth=0.5) # adding grid
plt.show()
```

xlim()  controls X-axis range , 
ylim()  controls Y-axis range

```python
month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.plot(month, budget, color='blue', linestyle='--', marker='o' ,linewidth=2, label= 'Monthly Budget')
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.title("Monthly Budget Over 4 Months")
plt.legend(loc='upper left')  # at the upper left corner
plt.grid(color='gray', linestyle=':', linewidth=0.5) # adding grid
plt.xlim(0, 5)  # setting x-axis limits
plt.ylim(2000, 3200)  # setting y-axis limits
plt.show()
```

xticks() & yticks() 

```python
month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.plot(month, budget, color='blue', linestyle='--', marker='o' ,linewidth=2, label= 'Monthly Budget')
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.title("Monthly Budget Over 4 Months")
plt.legend(loc='upper left')  # at the upper left corner
plt.grid(color='gray', linestyle=':', linewidth=0.5) # adding grid
plt.xlim(0, 5)  # setting x-axis limits
plt.ylim(2000, 3200)  # setting y-axis limits
plt.xticks([1, 2, 3, 4], ['Jan', 'Feb', 'Mar', 'Apr'])  # setting x-axis ticks
plt.show()
```

bar() chart

```python
# plt.bar(valueX, valueY, color='color_name', width=value, label='label name')

month = [1, 2, 3, 4]
budget = [2500, 2700, 3000, 2800]
plt.bar(month, budget, color='blue', width=0.4, label= 'Monthly Budget')
# plt.barh(month, budget, color='blue', width=0.4, label= 'Monthly Budget') # horizontal bar chart
plt.xlabel("Month")
plt.ylabel("Budget in $")
plt.legend(loc='upper left')  # at the upper left corner
plt.title("Monthly Budget Over 4 Months")
plt.show()
```

pie() chart

```python
# plt.pie(valueX, valueY, labels=labels_list, colors=['color1', 'color2', ...], autopct='format', startangle=value , shadow=True/False)

aqi = [500, 180 , 290 , 299]
cities = ['Gr.Noida', 'Gorakhpur', 'Lucknow', 'Motihari']
plt.pie(aqi, labels=cities, colors=['red', 'green', 'brown', 'yellow'], autopct='%1.1f%%', startangle=140 , shadow=True)
plt.title("Air Quality Index (AQI) Distribution")
plt.show()
```

hist() chart

```python
# plt.hist(valueX, bins=value, color='color_name', edgecolor='color_name', label='label name')

marks = [55, 67, 45, 78, 89, 90, 56, 77, 88, 92, 61, 73, 84, 69, 95]
plt.hist(marks, bins=5, color='blue', edgecolor='black', label='Student Marks Distribution')
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.legend(loc='upper left')
plt.title("Distribution of Student Marks")
plt.show()
```

scatter() -> important most useful in machine learnning

```python
#plt.scatter(valueX, valueY, color='color_name', marker='marker_symbol', label='label name')

hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 65, 70, 75, 80, 85, 90]
plt.scatter(hours_studied, scores, color='green', marker='o', label='Study Hours vs Scores')
plt.xlabel("Hours Studied")
plt.ylabel("Scores Obtained")
plt.legend()
plt.grid(color='gray', linestyle=':', linewidth=0.5) 
plt.title("Study Hours vs Scores Scatter Plot")
plt.show()
```

```python
plt.scatter([1,2,3,4], [50,55,65,70], color='green', marker='o', label='Class A Students')
plt.scatter([1,2,3,4], [40,50,55,80], color='blue', marker='x', label='Class B Students')
plt.xlabel("Hours Studied")
plt.ylabel("Scores Obtained")
plt.legend()
plt.grid(color='gray', linestyle=':', linewidth=0.5) 
plt.title("Comparison of Class A and Class B Students")
plt.show()
```
