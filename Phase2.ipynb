```python
import matplotlib.pyplot as plt
```

Subplot()

```python
# plt.subplot(nrows, ncols, index)

x = [1, 2, 3, 4]
y = [2500, 2700, 3000, 3100]
plt.subplot(1, 2, 1)  # 1 row, 2 column, first subplot
plt.plot(x, y, color='blue', marker='o', linestyle='-', label='Monthly Budget')
plt.xlabel("Months")
plt.ylabel("Budget in USD")
plt.title("Line Chart")
plt.legend(loc='upper left')  
plt.grid(color='gray', linestyle=':', linewidth=0.5) 
plt.subplot(1, 2, 2)  # 1 row, 2 column, second subplot
plt.bar(x, y, color='orange', label='Monthly Budget Bar')
plt.xlabel("Months of bar chart")
plt.ylabel("Budget in USD")
plt.title("Bar Chart")
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Adjust graph size :

```python
# fig,ax = plt.subplots(nrows1, ncols1, figsize=(width, height))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns
x = [1, 2, 3, 4]
y = [2500, 2700, 3000, 3100]

# First subplot: line chart
ax[0].plot(x, y, color='blue', marker='o', linestyle='-', label='Monthly Budget')
ax[0].set_title("Line Chart")
ax[0].set_xlabel("Months")
ax[0].set_ylabel("Budget in USD")
ax[0].legend(loc='upper left')
ax[0].grid(color='gray', linestyle=':', linewidth=0.5)

# Second subplot: bar chart
ax[1].bar(x, y, color='orange', label='Monthly Budget Bar')
ax[1].set_title("Bar Chart")
ax[1].set_xlabel("Months of bar chart")
ax[1].set_ylabel("Budget in USD")   
ax[1].legend(loc='upper left')

# Main title for the entire figure
fig.suptitle("Monthly Budget Analysis", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

Saving the charts :

```python
# savfig('fileName.format", dpi=resolution , bbox_inches='tight')
  # always save before plt.show()

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns
x = [1, 2, 3, 4]
y = [2500, 2700, 3000, 3100]

# First subplot: line chart
ax[0].plot(x, y, color='blue', marker='o', linestyle='-', label='Monthly Budget')
ax[0].set_title("Line Chart")
ax[0].set_xlabel("Months")
ax[0].set_ylabel("Budget in USD")
ax[0].legend(loc='upper left')
ax[0].grid(color='gray', linestyle=':', linewidth=0.5)

# Second subplot: bar chart
ax[1].bar(x, y, color='orange', label='Monthly Budget Bar')
ax[1].set_title("Bar Chart")
ax[1].set_xlabel("Months of bar chart")
ax[1].set_ylabel("Budget in USD")   
ax[1].legend(loc='upper left')

# Main title for the entire figure
fig.suptitle("Monthly Budget Analysis", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('monthly_budget_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
